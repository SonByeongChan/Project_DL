```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torch import nn, optim
from PIL import Image
from sklearn.preprocessing import LabelEncoder

# CSV 파일에서 데이터 로드
qwer = pd.read_csv(r'F:\fungi_clef_2025\images\FungiTastic-FewShot\processed\equalized_images_with_labels_encoded.csv')

# 픽셀 값과 레이블 설정
pixels = qwer.iloc[:, :-1].values  # 마지막 두 컬럼을 제외한 모든 열 (픽셀 데이터)
labels = qwer.iloc[:, -1].values    # 마지막 열 (인코딩된 레이블)

width, height = 64, 64  # 이미지 크기 설정

# 사용자 정의 데이터셋 클래스
class CustomImageDataset(Dataset):
    def __init__(self, pixel_data, labels, transform=None):
        self.pixel_data = pixel_data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.pixel_data)

    def __getitem__(self, idx):
        # 픽셀 정보를 배열로 변환하고 8비트로 변환
        pixel_array = self.pixel_data[idx].astype(np.uint16)  # uint16로 변환
        pixel_array = (pixel_array / 16).astype(np.uint8)  # 0~4095 범위를 0~255로 변환

        image = Image.fromarray(pixel_array.reshape((height, width)), mode='L')  # (H, W)로 변환

        label = self.labels[idx]

        # 데이터 증강 적용
        if self.transform:
            image = self.transform(image)

        return image, label

# 데이터 증강 정의
data_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.Resize((height, width)),  # 이미지 크기 조정
    transforms.ToTensor(),
])

# 데이터셋 및 데이터로더 생성
dataset = CustomImageDataset(pixel_data=pixels, labels=labels, transform=data_transforms)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# 간단한 CNN 모델 정의
class ComplexCNN(nn.Module):
    def __init__(self, num_classes):
        super(ComplexCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)  # 입력 채널 수를 1로 설정
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        self.pool = nn.MaxPool2d(2)
        self.dropout = nn.Dropout(p=0.5)  # 드롭아웃 추가

        # Fully connected layers
        self.fc1 = nn.Linear(128 * (height // 8) * (width // 8), 256)
        self.bn_fc1 = nn.BatchNorm1d(256)
        self.fc2 = nn.Linear(256, 128)
        self.bn_fc2 = nn.BatchNorm1d(128)
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = nn.functional.relu(self.bn1(self.conv1(x)))
        x = self.pool(x)
        x = nn.functional.relu(self.bn2(self.conv2(x)))
        x = self.pool(x)
        x = nn.functional.relu(self.bn3(self.conv3(x)))
        x = self.pool(x)

        x = x.view(x.size(0), -1)  # Flatten
        x = nn.functional.relu(self.bn_fc1(self.fc1(x)))
        x = self.dropout(x)  # 드롭아웃 적용
        x = nn.functional.relu(self.bn_fc2(self.fc2(x)))
        x = self.fc3(x)
        return x

# 모델 사용 예시
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# model = ComplexCNN(num_classes=2413).to(device)

# 모델 훈련
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ComplexCNN(num_classes=2413).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 훈련 루프
num_epochs = 50
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # 정확도 계산
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss / len(dataloader):.4f}, Accuracy: {accuracy:.2f}%')

print("훈련 완료!")
```
