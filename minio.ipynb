```python
import os
import cv2
import numpy as np

#   
data_dir = r'E:\fungi_clef_2025\images\FungiTastic-FewShot\train\300p'
output_dir = r'E:\fungi_clef_2025\images\FungiTastic-FewShot\processed'
output_csv = os.path.join(output_dir, 'equalized_images.csv')

#   
os.makedirs(output_dir, exist_ok=True)

# CSV  
first_image = True  #      

with open(output_csv, 'w') as f:
    for filename in os.listdir(data_dir):
        img_path = os.path.join(data_dir, filename)  #   
        print(f"Processing: {img_path}")  #     

        # OpenCV  
        image = cv2.imread(img_path)

        #    
        if image is not None:
            #   (64x64)
            resized_image = cv2.resize(image, (64, 64))

            #  
            gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

            #  1  
            flattened_image = gray_image.flatten()

            if first_image:
                #      
                header = [f'pixel_{j}' for j in range(flattened_image.size)]
                f.write(','.join(header) + '\n')
                first_image = False

            # 1  CSV  
            np.savetxt(f, flattened_image[np.newaxis], delimiter=',', fmt='%d')
            print(f"Saved pixel values for: {filename}")  #   
        else:
            print(f"Error loading image: {img_path}")

print(f"All processed images have been saved to: {output_csv}")
```

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torch import nn, optim
from PIL import Image
from sklearn.preprocessing import LabelEncoder

# CSV   
qwer = pd.read_csv(r'F:\fungi_clef_2025\images\FungiTastic-FewShot\processed\equalized_images_with_labels_encoded.csv')

#    
pixels = qwer.iloc[:, :-1].values  #       ( )
labels = qwer.iloc[:, -1].values    #   ( )

width, height = 64, 64  #   

#    
class CustomImageDataset(Dataset):
    def __init__(self, pixel_data, labels, transform=None):
        self.pixel_data = pixel_data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.pixel_data)

    def __getitem__(self, idx):
        #     8 
        pixel_array = self.pixel_data[idx].astype(np.uint16)  # uint16 
        pixel_array = (pixel_array / 16).astype(np.uint8)  # 0~4095  0~255 

        image = Image.fromarray(pixel_array.reshape((height, width)), mode='L')  # (H, W) 

        label = self.labels[idx]

        #   
        if self.transform:
            image = self.transform(image)

        return image, label

#   
data_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.Resize((height, width)),  #   
    transforms.ToTensor(),
])

#    
dataset = CustomImageDataset(pixel_data=pixels, labels=labels, transform=data_transforms)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

#  CNN  
class ComplexCNN(nn.Module):
    def __init__(self, num_classes):
        super(ComplexCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)  #    1 
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        self.pool = nn.MaxPool2d(2)
        self.dropout = nn.Dropout(p=0.5)  #  

        # Fully connected layers
        self.fc1 = nn.Linear(128 * (height // 8) * (width // 8), 256)
        self.bn_fc1 = nn.BatchNorm1d(256)
        self.fc2 = nn.Linear(256, 128)
        self.bn_fc2 = nn.BatchNorm1d(128)
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = nn.functional.relu(self.bn1(self.conv1(x)))
        x = self.pool(x)
        x = nn.functional.relu(self.bn2(self.conv2(x)))
        x = self.pool(x)
        x = nn.functional.relu(self.bn3(self.conv3(x)))
        x = self.pool(x)

        x = x.view(x.size(0), -1)  # Flatten
        x = nn.functional.relu(self.bn_fc1(self.fc1(x)))
        x = self.dropout(x)  #  
        x = nn.functional.relu(self.bn_fc2(self.fc2(x)))
        x = self.fc3(x)
        return x

#   
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# model = ComplexCNN(num_classes=2413).to(device)

#  
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ComplexCNN(num_classes=2413).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

#  
num_epochs = 50
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        #  
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss / len(dataloader):.4f}, Accuracy: {accuracy:.2f}%')

print(" !")
```
